## 3.1.3 3次関数f(x) = (x-2) x (x+2)を描く

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f(x):
    return (x-2)*x*(x+2)
```

```python
print(f(1))
```

```python
print(f(np.array([1,2,3])))
```

```python
plt.plot(x,f(x))
plt.show()
```

## 3.1.6 グラフを装飾する 

```python
#関数を定義
def f2(x,w):
    return (x-w)*x*(x+2)

#xを定義
x=np.linspace(-3,3,100)

#グラフを描写
plt.plot(x,f2(x,2),color='black',label='$w=2$')
plt.plot(x,f2(x,1),color='cornflowerblue', label='$w=1$')
plt.legend(loc='upper left') #凡例表示
plt.ylim(-15,15) #y軸の範囲
plt.title('$f_2(x)$') #タイトル
plt.xlabel('$x$') #xラベル
plt.ylabel('$y$') #yラベル
plt.grid(True) #グリッド
plt.show()
```

## 3.1.7 グラフを複数並べる

```python
plt.figure(figsize=(10,3))
plt.subplots_adjust(wspace=0.5,hspace=0.5)
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.title(i+1)
    plt.plot(x,f2(x,i),'k')
    plt.ylim(-20,20)
    plt.grid(True)
plt.show()
```

# 3.2 3次元のグラフを描く

## 3.2.1 2つの変数の関数

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f3(x0,x1):
    ans=(2*x0**2+x1**2)*np.exp(-(2*x0**2+x1**2))
    return ans

xn=9
x0=np.linspace(-2,2,xn)
x1=np.linspace(-2,2,xn)
y=np.zeros((len(x0),len(x1)))
for i0 in range(xn):
    for i1 in range(xn):
        y[i1,i0]=f3(x0[i0],x1[i1])
```

```python
print(x0)
```

```python
print(x1)
```

```python
print(y)
```

```python
print(np.round(y,1))
```

## 3.2.2 数値を色で表現する : pcolor

```python
plt.figure(figsize=(3.5,3))
plt.gray()
plt.pcolor(y)
plt.colorbar()
plt.show()
```

## 3.2.3 関数の表面を面で表す : surface

```python
from mpl_toolkits.mplot3d import Axes3D
xn=50
xx0,xx1=np.meshgrid(x0,x1)

plt.figure(figsize=(5,3.5))
ax=plt.subplot(1,1,1,projection='3d')
ax.plot_surface(xx0,xx1,y,rstride=1,cstride=1,alpha=0.3,color='blue',edgecolor='black')
ax.set_zticks((0,0.2))
ax.view_init(75,-95)
plt.show()
```

## 3.2.4 等高線で表示する : contour

```python
xn=50
x0=np.linspace(-2,2,xn)
x1=np.linspace(-2,2,xn)

y=np.zeros((xn,xn))
for i0 in range(xn):
    for i1 in range(xn):
        y[i1,i0]=f3(x0[i0],x1[i1])

xx0,xx1=np.meshgrid(x0,x1)

plt.figure(1,figsize=(4,4))
cont=plt.contour(xx0,xx1,y,5,colors='black')
cont.clabel(fmt='%.2f',fontsize=8)

plt.show()
```
